```python
import pandas as pd

df = pd.read_csv('../data/input/frota_detran_por_uf_2025.csv', sep=';')
df.head()
```

```python
df["Ano Fabricação Veículo CRV"] = pd.to_numeric(
    df["Ano Fabricação Veículo CRV"], errors="coerce"
)

df["Qtd. Veículos"] = pd.to_numeric(
    df["Qtd. Veículos"], errors="coerce"
)

df = df.dropna(subset=["Ano Fabricação Veículo CRV", "Qtd. Veículos"])

df["Ano Fabricação Veículo CRV"] = df["Ano Fabricação Veículo CRV"].astype(int)
df["Qtd. Veículos"] = df["Qtd. Veículos"].astype(int)
```

```python
df["IMPORTADO"] = df["Marca Modelo"].str.startswith("I/", na=False)

df["Marca Modelo s/ importado"] = df["Marca Modelo"].str.removeprefix("I/")
```

```python
df["qtd_barras"] = df["Marca Modelo"].str.count("/")

df.loc[df["qtd_barras"] >= 1, "MARCA"] = (
    df.loc[df["qtd_barras"] >= 1, "Marca Modelo s/ importado"]
      .str.split("/")
      .str[0]
)

df["Marca Modelo s/ marca"] = (
    df["Marca Modelo s/ importado"]
      .str.replace(r"^[^/]+/", "", regex=True)
)
```

```python
df_importados = df[df['IMPORTADO']]
df_importados.head()
```

```python
# Ordenar marcas por tamanho (maiores primeiro)
brands_sorted = sorted(brands, key=len, reverse=True)

def extrair_montadora(texto):
    for marca in brands_sorted:
        if texto.startswith(marca):
            return marca
    return None

df_importados['MARCA'] = df_importados['Marca Modelo s/ marca'].apply(extrair_montadora)
df_importados.head()
```

```python
import re

df_importados['Marca Modelo s/ marca'] = df_importados.apply(
    lambda row: re.sub(
        r'^' + re.escape(row['MARCA']) + r'\s*',
        '',
        row['Marca Modelo s/ marca']
    ) if pd.notna(row['MARCA']) else row['Marca Modelo s/ marca'],
    axis=1
)
```

```python
mask_imp = df['MARCA'] == 'IMP'

df.loc[mask_imp, 'IMPORTADO'] = True
```

```python
df.loc[mask_imp, 'MARCA'] = (
    df.loc[mask_imp, 'Marca Modelo s/ marca']
    .apply(extrair_montadora)
)
```

```python
df.loc[mask_imp, 'Marca Modelo s/ marca'] = (
    df.loc[mask_imp]
    .apply(
        lambda row: re.sub(
            r'^' + re.escape(row['MARCA']) + r'\s*',
            '',
            row['Marca Modelo s/ marca']
        ) if pd.notna(row['MARCA']) else row['Marca Modelo s/ marca'],
        axis=1
    )
)
```

```python
df['ARTESANAL'] = df['MARCA'].str.startswith('A.', na=False)
```

```python
df.dropna(subset=['MARCA'], inplace=True)
```

```python
resumo = df[(~df['ARTESANAL'])].groupby("MARCA")["Qtd. Veículos"].sum()
marcas_validas = resumo[resumo > 10]

df_final = df[df['MARCA'].isin(marcas_validas.index)]
df_final = df_final.dropna(subset=['MARCA'])
df_final.head()
```
